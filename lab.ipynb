```python
def f(x):
    return 0.06*x**6 - 0.016*x**5 + 0.04*x**4 + 0.010*x**3 - 0.03*x**2 - 0.05*x - 0.012
```

```python
import numpy as np
import matplotlib.pyplot as plt
x_values = np.linspace(-1, 1, 1000)
y_values = f(x_values)
plt.plot(x_values, y_values)
plt.title('Đồ thị của hàm số')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from scipy.optimize import root_scalar
import numpy as np
import time

start_time = time.time()
result_bisect = root_scalar(f, method='bisect', bracket=[-1, 0.5])
end_time = time.time()
print("Nghiệm bisect:", result_bisect.root)
print("Thời gian bisect:", end_time - start_time, "giây")

start_time = time.time()
result_newton = root_scalar(f, method='newton', x0=0)
end_time = time.time()
print("Nghiệm Newton:", result_newton.root)
print("Thời gian Newton:", end_time - start_time, "giây")

start_time = time.time()
result_secant = root_scalar(f, method='secant', x0=-3, x1=3)
end_time = time.time()
print("Nghiệm Secant:", result_secant.root)
print("Thời gian Secant:", end_time - start_time, "giây")
```

```python
def fm(x1, x2):
    x = np.array([x1, x2])
    A = np.array([[4, -6], [-3, 7]])
    c = np.array([[5], [-6]])
    d = 10
    return x.T@A@x + c.T@x + d
```

```python
fm(2,3)
```

```python
from scipy.optimize import minimize
def objective_function(x):
    return fm(x)

# Giả sử điểm bắt đầu
x0 = np.array([0, 0])

# Tìm giá trị tối ưu với phương pháp Newton-CG
result = minimize(objective_function, x0, method='Newton-CG')
print("Nghiệm tối ưu:", result.x)
print("Giá trị tối ưu:", result.fun)
```
